```python
plt.figure(1)
plt.plot(y, np.sin(y))
plt.figure(2)
plt.plot(y, np.cos(y))
plt.savefig('np.cosine.png')
plt.figure(1)
plt.title('np.sin(y)')
plt.savefig('np.sine.png')
```
